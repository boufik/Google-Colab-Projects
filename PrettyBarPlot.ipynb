# AUXILIARY FUNCTION

```python
import matplotlib.pyplot as plt

def pretty_barplot(X_data, Y_data, custom_color):
  plt.figure()
  bars = plt.bar(X_data, Y_data, color=custom_color)
  plt.xlabel("Round")
  plt.ylabel("Value")
  plt.title("Average value per round")
  # Add centered value labels inside each bar
  for bar in bars:
      height = bar.get_height()
      plt.text(bar.get_x() + bar.get_width() / 2, height / 2, f"{100*height:.2f}%", ha="center", va="center")
  plt.tight_layout()
  plt.show()
```

# MAIN FUNCTION

```python
Y_data = [0.1229, 0.1896, 0.2457, 0.3779, 0.5321, 0.7187, 0.9079]
X_data = list(range(1, len(Y_data) + 1))
custom_color = "#a8dae8"
pretty_barplot(X_data, Y_data, custom_color)
```
